```python
basket = data.groupby(['Project_Name','month'])['Item_Name'].apply(list).reset_index()
basket
```

```python
from mlxtend.frequent_patterns import association_rules, apriori
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_ary = te.fit(basket['Item_Name']).transform(basket['Item_Name'])
basket_encoded = pd.DataFrame(te_ary,columns = te.columns_)
basket_encoded
```

```python
freq_items = apriori(basket_encoded,min_support=0.1,use_colnames=True)
freq_items
```

```python
rules = association_rules(freq_items,metric = 'lift',min_threshold = 2)
rules
```

```python
rules.sort_values(by = 'lift',ascending = False,inplace = True)
rules
```

```python
rules.to_csv('Association_rules_requisition.csv')
```
