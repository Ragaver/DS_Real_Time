```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
import datetime
new_date = datetime.datetime(2024,9,6)
```

```python
RFM_approach = data.groupby('Supplier_Name').agg({'Date':lambda x : (new_date - x.max()).days,
                                                  'PO_ID':lambda x : x.count(),
                                                  'Value_INR': lambda x: x.sum()},inplace = True)
RFM_approach.rename(columns = {'Supplier_Name':'Supplier','Date':'Recency','PO_ID':'Frequency','Value_INR':'Monitary'},inplace = True)
RFM_approach.reset_index()
```

```python
quantiles = RFM_approach.quantile(q = [0.20])
quantiles = quantiles.to_dict()
quantiles
```

```python
def recency_val(x,p,d):
    if x<=d[p][0.20]:
        return 1
    else:
        return 2
```

```python
def freq_moni_val(x,p,d):
    if x<=d[p][0.20]:
        return 2
    else:
        return 1
```

```python
RFM_approach['R'] = RFM_approach['Recency'].apply(recency_val,args = ('Recency',quantiles))
RFM_approach['F'] = RFM_approach['Frequency'].apply(freq_moni_val,args = ('Frequency',quantiles))
RFM_approach['M'] = RFM_approach['Monitary'].apply(freq_moni_val,args = ('Monitary',quantiles))
RFM_approach = RFM_approach.reset_index()
```

```python
RFM_approach
```

```python
RFM_approach['RFM'] = RFM_approach[['R','F','M']].sum(axis = 1)
```

```python
RFM_approach['RFM'].nunique()
```

```python
spend_level = ['20%','80%']
spend_cut = pd.qcut(RFM_approach['RFM'],q = 2,labels = spend_level)
RFM_approach['Category'] = spend_cut.values
RFM_approach.reset_index()
```

```python
RFM_approach.to_csv('RFM_Pareto_combined_approach.csv')
```

```python
final_report = RFM_approach.groupby('Category')[['Recency','Frequency','Monitary','Supplier_Name']].agg(
    {'Recency':'sum','Frequency':'sum','Monitary':'sum','Supplier_Name':'nunique'}).reset_index()
final_report
```

```python
final_rep = final_report[['Category','Supplier_Name']]
```

```python
final_rep['percent'] = final_report['Supplier_Name']/sum(final_report['Supplier_Name'])*100
```

```python
final_rep.rename(columns = {'Supplier_Name':'Supplier_count','percent':'Supplier_percent'},inplace = True)
```

```python
final_rep
```
